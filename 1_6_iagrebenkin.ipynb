```python
import numpy as np
```

```python
data = np.zeros((10**5,3))
```

```python
data[:, 0] = np.random.normal(400000, 500, 10**5)
data[:, 1] = np.random.normal(300000, 1000, 10**5)
data[:, 2] = np.random.normal(3.20, 0.01, 10**5)
```

```python
Fs = np.zeros(10**5)
G = 6.67 * 10**(-11)

Fs[:] = G * data[:, 0] * data[:, 1] / data[:, 2] ** 2
```

```python
import matplotlib.mlab as mlab        
import matplotlib.pyplot as plt
from scipy.stats import norm


plt.hist(Fs)
mu = 0.7816
sigma = 6*10**(-3)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, 500*scipy.stats.norm.pdf(x, mu, sigma))
plt.show()
```
